```python

import pandas as pd
import os
# from bs4 import BeautifulSoup
```

```python
etfFolder="../00ref/成分股/"
shzsTarget = etfFolder+"上证50ETF.xlsx"
shzs_df = pd.read_excel(shzsTarget)
shzs_df = shzs_df[['股票代码', '股票简称']].dropna()
shzs_df.iloc[:,0] = shzs_df.iloc[:,0].astype(int).astype(str)
shzs_s = shzs_df.iloc[:,1]
shzs_s.head()
```

```python
SZSE_Target = etfFolder+"深证100ETF.xls"
SZSE_df = pd.read_excel(SZSE_Target,dtype=str)
SZSE_df = SZSE_df[["证券代码", "证券简称"]].dropna()
SZSE_s = SZSE_df.iloc[:,1]
SZSE_s.head()
```

```python
ROOT = "../03featureExtraction_featureSelection/news_iosips/"
thisFile = os.listdir(ROOT)[-1]
target = ROOT+"{}".format(thisFile)
print(target)
df = pd.read_pickle(target)
content_s = df.content
df
```

```python
# aStock = shzs_s.iloc[0]
companyDF_dict = dict()

for aStock in shzs_s:
    companyDF_dict.update({aStock:df.loc[df.shzs_company_name.apply(lambda x: aStock in x)]})
    
for aStock in SZSE_s:
    companyDF_dict.update({aStock:df.loc[df.szse_company_name.apply(lambda x: aStock in x)]})
```

```python
aCompanyTable = companyDF_dict['中国重工'].set_index('datetime')
aCompanyTable.index = pd.DatetimeIndex(aCompanyTable.index)
aResampledTable = aCompanyTable.resample('M').sum()
```

```python
companyIOSIPS_dict = {}
for aCompany,aTable in companyDF_dict.items():
    aCompanyTable = aTable.set_index('datetime')
    aCompanyTable.index = pd.DatetimeIndex(aCompanyTable.index)
    aIOSIPSTable = aCompanyTable.resample('M').sum().loc[:,['IOS_news','IPS_news']]
    companyIOSIPS_dict.update({aCompany:aIOSIPSTable})
```
