```python
import pandas as pd
import numpy as np
import os, pickle
```

```python
ROOT = "../03featureExtraction_featureSelection/news_scored/"
thisFile = os.listdir(ROOT)[-1]
target = ROOT+"{}".format(thisFile)
print(target)
df = pd.read_pickle(target)
content_s = df.content
df['label'] = 0
df = df.reset_index(drop = True)
df
```

```python
ROOT = "../03featureExtraction_featureSelection/vectorizer_model/"
thisFile = os.listdir(ROOT)[-1]
target = ROOT+"{}".format(thisFile)
print(target)
with open(target, 'rb') as file:
    vectorizer =pickle.load(file)
    
idf = vectorizer.idf_
term = vectorizer.get_feature_names()
idfDict = dict(zip(term, idf))
```

## labeling 

```python
superGoodNews = df.sort_values(by = 'hybridSentimentScore',ascending=False)[:int(len(df)*0.05)]
df.loc[superGoodNews.index,'label'] = 1
superBadNews = df.sort_values(by = 'hybridSentimentScore',ascending=True)[:int(len(df)*0.05)]
df.loc[superBadNews.index,'label'] = -1
df.label.value_counts()
```

```python
X = vectorizer.transform(df.joinedTokens_s)
y = df.label
print(X.shape)
```

## to bool

```python
X = X>0
```

# selection
## select posterm

```python
from sklearn.feature_selection import SelectKBest, SelectPercentile, SelectFpr
from sklearn.feature_selection import chi2, mutual_info_classif, f_classif
```

```python
select_function = chi2
alpha = 0.01
```

```python
y_pos = (y>0)

posSelector = SelectFpr(select_function, alpha=alpha)
posSelector.fit(X, y_pos)
X_pos = posSelector.transform(X)
print(X_pos.shape)
selectedArgs_pos = posSelector.get_support()

superGoodTerms = np.array(term)[selectedArgs_pos]
for aterm in superGoodTerms:
    print(aterm,idfDict[aterm])
```

```python
y_neg = (y<0)


negSelecter = SelectFpr(select_function, alpha=alpha)
negSelecter.fit(X, y_neg)
X_neg = negSelecter.transform(X)
print(X_neg.shape)
selectedArgs_neg = negSelecter.get_support()
superBadTerms = np.array(term)[selectedArgs_neg]

for aterm in superBadTerms:
    print(aterm,idfDict[aterm])
```

```python
IOS_news = X[:,selectedArgs_pos].sum(axis=1)
```

```python
IPS_news = X[:,selectedArgs_neg].sum(axis=1)
```

```python
from datetime import date
df['IOS_news'] = IOS_news
df['IPS_news'] = IPS_news
df.to_pickle('./news_iosips/news_iosips_{}'.format(date.today().strftime("%Y%m%d")))
```
