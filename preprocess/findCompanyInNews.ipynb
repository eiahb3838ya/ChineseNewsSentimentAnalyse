```python
import pandas as pd
import requests, os
from bs4 import BeautifulSoup
```

```python
ROOT = "../crawler/news_data/"
print(os.listdir(ROOT))
thisFile = "financeChina4"
target = ROOT+"{}.csv".format(thisFile)
```

```python
rawData = pd.read_csv(target)
rawData = rawData.rename(columns = {'Unnamed: 0':"id"})
```

```python
df = rawData.dropna(subset=['content','title'])
len(df.content)
```

# 上證50

```python
etfFolder="../ref/"
shzsTarget = etfFolder+"上证50ETF.xlsx"
```

```python
shzs_df = pd.read_excel(shzsTarget)
shzs_df = shzs_df[['股票代码', '股票简称']].dropna()
shzs_df.iloc[:,0] = shzs_df.iloc[:,0].astype(int).astype(str)
shzs_df.head()
```

```python
shzsCompanyName = shzs_df.iloc[:,1]
shzsCompanyCode = shzs_df.iloc[:,0]
```

```python
alist = []
for index, aNews in df.iterrows():
    aContent = aNews.content
    try:
        conpanyIsIn = shzsCompanyName.apply(lambda x: aContent.find(x))>0
        conpanyCodeIsIn = shzsCompanyCode.apply(lambda x: aContent.find(x))>0
    except Exception as e :
        print(str(e))
        print(index)
    conpanyInContent = shzsCompanyName.loc[conpanyIsIn|conpanyCodeIsIn]
    alist.append(list(conpanyInContent))
    if list(conpanyInContent):
        print(index,list(conpanyInContent))
df['shzs_company_name'] = alist
```

```python
print(len(df))
print(sum(df.shzs_company_name.apply(lambda x : len(x)>0)))
```
