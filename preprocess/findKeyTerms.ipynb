```python
rawData = pd.read_pickle(target)
df = rawData
df.head()
```

```python
etfFolder="../ref/"
shzsTarget = etfFolder+"key_terms_s.xlsx"
shzs_df = pd.read_excel(shzsTarget)
neg_terms = shzs_df.neg.unique()
pos_terms = shzs_df.pos.unique()
neg_terms = pd.Series([term[1:-1] for term in neg_terms])
pos_terms = pd.Series([term[1:-1] for term in pos_terms])
print("we have pos terms:{} \nwe have neg terms:{}".format(len(pos_terms),len(neg_terms)))
```

```python
alist = []
for index, aNews in df.iterrows():
    aContent = aNews.content
    try:
        PosIsIn = pos_terms.apply(lambda x: aContent.find(x))>0
    except Exception as e :
        print(str(e))
        print(index)
    posInContent = pos_terms.loc[PosIsIn]
    alist.append(list(posInContent))
    if list(posInContent):
        print(index,list(posInContent))
df['pos_terms'] = alist
```

```python
alist = []
for index, aNews in df.iterrows():
    aContent = aNews.content
    try:
        negIsIn = neg_terms.apply(lambda x: aContent.find(x))>0
    except Exception as e :
        print(str(e))
        print(index)
    negInContent = neg_terms.loc[negIsIn]
    alist.append(list(negInContent))
    if list(negInContent):
        print(index,list(negInContent))
df['neg_terms'] = alist
```
